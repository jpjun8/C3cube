```python
df = pd.read_csv('./1인당 관광수입.csv', encoding='cp949')
df2 = pd.read_csv('./방한 외래관광객 추이.csv', encoding='cp949')
df2 = df2.rename(columns = {'기준년월':'기준연월'})
df3 = df.merge(df2, on='기준연월')
df4 = df3.drop(columns = ["인원수"])
df4.columns = ["기준연월", "관광수입", "1인당 관광수입", "방한 외래관광객", "환율", "단위", "최고환율", "최저환율", "국제유가", "최고 국제유가", "최저 국제유가"]
df5 = df4.drop(columns = ["단위", "최고환율", "최저환율", "최고 국제유가", "최저 국제유가"])
df5.head()
```
