# Warm Up

```python
# @title 프로그램 11-1 : 데이터 가져오기
import pandas as pd
df = pd.read_csv("train.csv")
df
```

```python
# @title 프로그램 11-2 : 'datetime' 열을 날짜 형태로 하며 데이터 가져오기
train = pd.read_csv("train.csv", parse_dates = ['datetime'])
# df.info()
train.info()
```

```python
# @title 프로그램 11-3 : 데이터 프레임에서 시간 추출하기
train["year"] = train["datetime"].dt.year
train["month"] = train["datetime"].dt.month
train["day"] = train["datetime"].dt.day
train["hour"] = train["datetime"].dt.hour
train["minute"] = train["datetime"].dt.minute
train["second"] = train["datetime"].dt.second
train["dayofweek"] = train["datetime"].dt.dayofweek
train
```

# Draw Code

```python
# @title season/weather, holiday/workingday 산점도 꺾은선 그래프 (시간 기준)

# 계절별
sns.pointplot(data = train, x = "hour", y = "count", hue = "season")
```

- 겨울이 가장 대여 횟수가 적고, 그 외에는 비슷. 게절 불문 출퇴근 시간에 가장 많음

```python
# 날씨별
sns.pointplot(data = train, x = "hour", y = "count", hue = "weather")
```

- 4번처럼 날씨가 몹시 안 좋은 날에는 대여가 거의 없다. 
- 3번도 1, 2번처럼 좋은 날씨보다는 확실히 대여 횟수가 적음.

```python
# 공휴일별
sns.pointplot(data = train, x = "hour", y = "count", hue = "holiday")
```

- 0번 (공휴일이 아님) : 출퇴근 시간대에 대여율이 높음
- 1번 (공휴일) : 11시부터 18시까지 대여를 많이 함 -- 나들이나 운동 목적일거라 추측

```python
# 근무일
sns.pointplot(data = train, x = "hour", y = "count", hue = "workingday")
```

- 0번 (근무일이 아님) : 11시~18시에 가장 많이 빌림.
- 1번 (근무일) : 출퇴근 시간에 가장 많이 빌림.
