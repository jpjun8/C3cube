```python
import numpy as np
import pandas as pd

df = pd.read_csv("./kdrama.csv")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
tfidf_vec = tfidf.fit_transform(df['Synopsis'])
```

```python
from sklearn.metrics.pairwise import linear_kernel
cosine_similarities = linear_kernel(tfidf_vec)
```

```python
def get_recommendations(name, sim=cosine_similarities):
    index = df[df['Name'] == name].index[0]
    similarity = enumerate(sim[index])
    scores = sorted(similarity, key=lambda x: x[1], reverse=True)
    return df.iloc[[i for (i, t) in scores[1:11]]]
```

```python
get_recommendations("My Mister")
```
